```python
from numpy.random import rand
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = sp.Symbol('x')
y = sp.Symbol('y')
```

```python
from sympy import exp
```

```python
sp.integrate( 1/10*exp(-y/10), (y, 0, x) )
```

```python
xx = np.linspace(0, 40, 1000)
yy = 1 - 1*np.exp(-xx/10)
plt.plot(xx, yy, 'b-')
```
